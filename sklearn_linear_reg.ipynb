```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
input_feature = np.linspace(0 , 10 , 100).reshape(-1 , 1)
final_output =  input_feature **2  + 1+ np.random.rand(100 , 1)* 20
```

```python
plt.scatter(input_feature , final_output , color = 'grey')
plt.show()
```

```python
# degee 1

model = LinearRegression()

X , X_test , y , y_test = train_test_split(input_feature, final_output
                                                       , test_size = 0.1
                                                       , random_state = 42)
X_train , X_val , y_train , y_val = train_test_split(X, y
                                                       , test_size = 0.1
                                                       , random_state = 42)

model.fit(X_train , y_train)

print('slope : ', model.coef_ ,'inercept : ' , model.intercept_)

y_pred_val = model.predict(X_val )

y_pred_train = model.predict(X_train )

y_pred_test = model.predict(X_test )


validation_error = mean_squared_error( y_val , y_pred_val  )
train_error = mean_squared_error( y_train , y_pred_train  )
test_error = mean_squared_error( y_test , y_pred_test )

plt.scatter(X_train , y_train , color = 'pink')
plt.scatter(X_val , y_val , color = 'yellow')
plt.scatter(X_test , y_test , color = 'skyblue')
plt.plot( X_train , y_pred_train , color = 'red')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
# degee 2

poly2 = PolynomialFeatures(degree = 2)
X_train_poly2 = poly2.fit_transform( X_train )
X_val_poly2 = poly2.fit_transform( X_val )
X_test_poly2 = poly2.fit_transform( X_test )

model2 = LinearRegression()

model2.fit( X_train_poly2 , y_train)

print('slope : ', model2.coef_ ,'inercept : ' , model2.intercept_)

y_pred_val = model2.predict( X_val_poly2 )

y_pred_train = model2.predict(X_train_poly2 )

y_pred_test = model2.predict( X_test_poly2 )


validation_error = mean_squared_error( y_val , y_pred_val  )
train_error = mean_squared_error( y_train , y_pred_train  )
test_error = mean_squared_error( y_test , y_pred_test )


print('validation_error:' ,  validation_error)
print('train_error :' , train_error)
print('test_error :' ,test_error)


plt.scatter(X_train , y_train , color = 'pink')
plt.scatter(X_val , y_val , color = 'yellow')
plt.scatter(X_test , y_test , color = 'skyblue')

# Generate points for plotting the curve fitting over the entire range of input_feature
X_plot = np.linspace(input_feature.min(), input_feature.max(), 100).reshape(-1, 1)
X_plot_poly2 = poly2.fit_transform(X_plot)
y_plot_pred = model2.predict(X_plot_poly2)

plt.plot( X_plot , y_plot_pred , color = 'red')
plt.show()
```
